# OPerating on Null value

```python
# isnull = check for null value
# notnull = opposite of isnull
# dropna = return the filtered dataset with null removed
# fillna = fill missing data with a specified value and returns a copy
```

```python
df_biodata.head()
```

```python
# check missing data
df_biodata.isna().head()
```

### Removing null values from pd series using notnull

```python
# Using notnull with a pandas series
df_biodata_series = df_biodata['Salary'] # extract a series from dataframe
df_biodata_series = df_biodata_series.head(10) # take out 10 rows from the series
df_biodata_series = df_biodata_series[df_biodata_series.notnull()] #remove the null values from the series
df_biodata_series.head()
```

### Removing null values from pd dataframe using notnull

```python
# Using notnull with a pandas series
df_biodata_frame = df_biodata.head(10) # take out 10 rows from the series
df_biodata_frame = df_biodata[df_biodata.notnull()] #remove the null values from the series
df_biodata_frame.head()
## This doesn't work because notnul returns a full dataframe of boolean rather than a boolean series that can be used to filter row by row
```

### How to remove NaN from dataframe using dropna

```python
df_biodata_drop = df_biodata.head(10)
# df_biodata_drop = df_biodata_drop.dropna() # drops every row with a NaN value
df_biodata_drop = df_biodata_drop.dropna(axis=1) # drops every column with a NaN value 
# how = .any = any column with NaN 
df_biodata_drop.head()
```

```python
# Filter rows where both Salary and Age are not null
df_not = df_biodata[['Salary', 'Age']].notnull()
df_biodata[df_not.all(axis=1)]
# Filter rows where both Salary or Age are not null
df_biodata[df_not.any(axis=1)]  
```
