## removing outliers

```python
# removing outliers using IQR interquotal range
Q1 = df_covid['New cases'].quantile(0.25)
Q3 = df_covid['New cases'].quantile(0.75)
IQR = Q3 - Q1
```

```python
# define bounds for acceotable data point
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q1 + 1.5 * IQR
```

```python
#filter out outlier
df_covid_no_outliers = df_covid[(df_covid['New cases'] >= lower_bound) & (df_covid['New cases'] <= upper_bound)]
```

### common aggregation function
| Function  | Description                    |
|-----------|--------------------------------|
| `sum()`   | Sum of values in each group.   |

| `max()`   | Maximum value.                |

```python
# create a numpy array from ocd column
import numpy as np
np_ocd = np.array(df_ocd['Duration of Symptoms (months)'])
```

```python
np_ocd
```

```python
def introduce_missing_values(data_column, missing_percentage=0.1):
    # make array a float type to accomodate NAN
    data_column = data_column.astype(float)
    
    
    num_records = len(data_column)  # Number of records in the column
    n_missing = int(num_records * missing_percentage)  # Number of missing values to introduce

    # Randomly choose indices to be replaced by NaN
    missing_indices = np.random.choice(num_records, n_missing, replace=False)

    # Set the chosen indices to NaN
    data_column[missing_indices] = np.nan

    return data_column
```

```python
np_ocd_NAN = introduce_missing_values(np_ocd)
```
