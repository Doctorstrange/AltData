#### PROBABILITY

```python
import matplotlib.pyplot as plt
```

```python
#PROBABILITY is a measure of chance. lets start from facts.
#The prrobability of the sun rising in the morning is 1 or 100, this is a definite truth. 
#The probability of a cat barking is 0 - impossibile
```

```python
#Example 1: when tossing a coin, you can get either head or tails, 
#this makes it a 50-50 chance of head or tails occuring in lay terms, the exact measure of this chance is probability
# Probability = number of favorable outcome/n   , n= number of possible outcomes 
# solution ; 1 (number of favourable outcome)/2(total number of possible outcomes) ; 1/2 is the probabilty of head occuring in a coin toss
```

```python
#Example 2: The probability of an outcome when rolling a dice
#using the above formular, Dice has 6 possible outcomes which is 1-6, 
# to calculate the probability of any of the numbers occuring is calculated thus:
# 1/6 (1 is the number of any favourable outcomes because a dice has one of each number)
# Example 3: what if a dice had the same number in 2 sides- then the probability of tht number occuring would be; 2/6 
```

```python
# the probabilty can be converted to a percentage by multipying the resulting fraction by 100
# on that note; 
#Example 1 :
prob_coin_percentage = round(1/2* 100, 2)
print(f"The probabilty of a head outcome in a coin toss is  {prob_coin_percentage}%") 
#Example 2 : 
prob_six_percentage = round(1/6*100,2)
print(f"The probabilty of throwing a 6 in a dice roll is  {prob_six_percentage}%")  
```

```python
labels = ['Heads in a Coin Toss', 'Throwing a 6 in Dice Roll'] 
probabilities = [prob_coin_percentage, prob_six_percentage] 
plt.bar(labels, probabilities, color=['blue', 'green'])
plt.xlabel('Event')
plt.ylabel('Probability (%)') 
plt.title('Probability of Events')
plt.ylim(0, 100);
```
